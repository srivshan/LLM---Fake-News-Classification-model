```python
import pandas as pd

# Load your dataset
df = pd.read_csv("output.csv")

# Check class distribution
print(df["label"].value_counts())  
```

```python

df_fake = df[df["label"] == 0]
df_real = df[df["label"] == 1]


min_samples = min(len(df_fake), len(df_real))

df_fake_balanced = df_fake.sample(min_samples, random_state=42)
df_real_balanced = df_real.sample(min_samples, random_state=42)


df_balanced = pd.concat([df_fake_balanced, df_real_balanced])

df_balanced = df_balanced.sample(frac=1, random_state=42).reset_index(drop=True)

print(df_balanced["label"].value_counts())  # Check if balanced
```

```python
from sklearn.utils import resample

df_fake_oversampled = resample(df_fake, replace=True, n_samples=len(df_real), random_state=42)


df_balanced = pd.concat([df_fake_oversampled, df_real])

df_balanced = df_balanced.sample(frac=1, random_state=42).reset_index(drop=True)

print(df_balanced["label"].value_counts())  # Check if balanced
```

```python
from sklearn.model_selection import train_test_split


train_df, test_df = train_test_split(df_balanced, test_size=0.2, random_state=42, stratify=df_balanced["label"])

print(train_df["label"].value_counts())  # Check balance in train set
print(test_df["label"].value_counts())   # Check balance in test set
```

```python
from transformers import RobertaTokenizer, RobertaForSequenceClassification

# Load the tokenizer and model
model_name = "roberta-base"  # or "roberta-large"
tokenizer = RobertaTokenizer.from_pretrained(model_name)
model = RobertaForSequenceClassification.from_pretrained(model_name, num_labels=2)
```

```python
train_df["combined_text"] = train_df["title"]  
```

```python
train_texts, val_texts, train_labels, val_labels = train_test_split(
    train_df["combined_text"], train_df["label"], test_size=0.2, random_state=42, stratify=train_df["label"]
)
```

```python
from datasets import Dataset


train_data = Dataset.from_dict({"text": train_texts, "label": train_labels})
val_data = Dataset.from_dict({"text": val_texts, "label": val_labels})


def tokenize_function(examples):
    return tokenizer(examples["text"], padding="max_length", truncation=True, max_length=512)

train_data = train_data.map(tokenize_function, batched=True)
val_data = val_data.map(tokenize_function, batched=True)
```

```python
from transformers import TrainingArguments, Trainer

training_args = TrainingArguments(
    output_dir="./results",
    eval_strategy="epoch",
    save_strategy="epoch",
    per_device_train_batch_size=8,
    per_device_eval_batch_size=8,
    num_train_epochs=3,
    weight_decay=0.01,
    logging_dir="./logs",
    logging_steps=10,
    load_best_model_at_end=True,
    report_to = 'tensorboard'
)

trainer = Trainer(
    model=model,
    args=training_args,
    train_dataset=train_data,
    eval_dataset=val_data
)

trainer.train()
```

```python
from sklearn.metrics import accuracy_score, precision_recall_fscore_support

def compute_metrics(eval_pred):
    predictions, labels = eval_pred
    preds = predictions.argmax(axis=1)
    acc = accuracy_score(labels, preds)
    precision, recall, f1, _ = precision_recall_fscore_support(labels, preds, average="binary")
    return {"accuracy": acc, "precision": precision, "recall": recall, "f1": f1}

trainer.evaluate()
```

```python
model.save_pretrained("fake-news-bert")
tokenizer.save_pretrained("fake-news-bert")
```
